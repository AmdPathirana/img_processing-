```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
(x_train, _), (_, _) = tf.keras.datasets.cifar10.load_data()
# Normalize pixel values to between -1 and 1
x_train = (x_train.astype('float32') - 127.5) / 127.5

# Generator model
def build_generator():
    model = models.Sequential()
    model.add(layers.Dense(4*4*256, use_bias=False, input_shape=(100,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((4, 4, 256)))

    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(3, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))

    return model

# Generator and discriminator
generator = build_generator()

# Generate 10 GAN images
num_images = 10
noise = tf.random.normal([num_images, 100])
generated_images = generator(noise, training=False)

# Display 10 GAN images
plt.figure(figsize=(20, 2))
for i in range(num_images):
    plt.subplot(1, num_images, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow((generated_images[i] + 1) / 2)  # Scale back to [0, 1] for display
plt.suptitle('GAN-generated Images')
plt.show()

# Display 10 original images
plt.figure(figsize=(20, 2))
for i in range(num_images):
    plt.subplot(1, num_images, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[np.random.randint(0, x_train.shape[0])])
plt.suptitle('Original Images')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the generator model
def build_generator():
    model = models.Sequential()
    model.add(layers.Dense(4*4*256, use_bias=False, input_shape=(100,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Reshape((4, 4, 256)))

    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(3, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))

    return model

# Load the trained generator
generator = build_generator()
generator.load_weights('generator_weights.h5')  # Load the weights of the trained generator

# Load the input image
input_image_path = 'truck.png'  # Provide the path to the input image
input_image = Image.open(input_image_path)
input_image = np.array(input_image)
input_image = (input_image.astype('float32') - 127.5) / 127.5  # Normalize pixel values to between -1 and 1
input_image = input_image[np.newaxis, :, :, :]  # Add batch dimension

# Generate an image using the generator
generated_image = generator(tf.random.normal([1, 100]), training=False)  # Generate an image from random noise

# Display the original and generated images
plt.figure(figsize=(8, 4))

# Display the original input image
plt.subplot(1, 2, 1)
plt.imshow((input_image[0] + 1) / 2)  # Scale back to [0, 1] for display
plt.title('Original Image')
plt.axis('off')

# Display the GAN-generated image
plt.subplot(1, 2, 2)
plt.imshow((generated_image[0] + 1) / 2)  # Scale back to [0, 1] for display
plt.title('GAN-generated Image')
plt.axis('off')

plt.show()
```
